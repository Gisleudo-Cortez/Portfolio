```python
# Create a validate_user() function, using some helper validation functions to validate user input. 

def validate_user(name, email, password):
    if not validate_name(name):
        raise ValueError("Invalid user name")
        return False
    if not validate_email(email):
        raise ValueError("Invalid user email")
        return False
    if not validate_password(password):
        raise ValueError("Invalid user password")
        return False
    return True
```

```python
# Testing 

print(validate_user('JohnDoe', 'johndoe@example.com', 'P@ssw0rd1'))  # Should Pass
try: 
    validate_user('', 'invalid@email', 'weakpassword') # Invalid cases - should raise ValueError
except ValueError as e:
    print(f"Caught the expected error: {e}")

try: 
    validate_user('John Doe', 'johndoe.com', '123456')  # More invalid cases
except ValueError as e:
    print(f"Caught the expected error: {e}")
```

```python
#  Create a register_user() function that registers the user if all checks pass. 

def register_user(name, email, password):
    if validate_user(name, email, password):
        out = {'name':name, 'email':email, 'password':password}
        return out
    return False
```

```python
# Testing the register_user function
print(register_user('JohnDoe', 'johndoe@example.com', 'P@ssw0rd1'))  # Should return a dictionary with valid data

try:
    register_user('', 'invalid@email', 'weakpassword') # Invalid cases - should raise ValueError, but we're testing register_user here
except ValueError as e:
    print(f"Caught the expected error in validate_user: {e}")

try: 
    register_user('John Doe', 'johndoe.com', '123456')  # More invalid cases - should trigger ValueError within validate_user
except ValueError as e:
    print(f"Caught the expected error in validate_user: {e}")
```
